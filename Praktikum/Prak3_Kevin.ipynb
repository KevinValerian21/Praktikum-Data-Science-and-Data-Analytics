```python
#Memasang library
import sys #Modul sys berisi fungsi yang berkaitan dengan interpreter Python dan lingkungannya, antara lain system.
!{sys.executable} -m pip install requests
!{sys.executable} -m pip install lxml
import requests
import lxml.html as lh
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#Scrape Table
url = 'http://pokemondb.net/pokedex/all'
page = requests.get(url)
doc = lh.fromstring(page.content)
tr_elements = doc.xpath('//tr')
```

```python
#Mengetahui jumlah kolom tabel yang telah dikikis
[len(Z) for Z in tr_elements[:8]]
```

```python
#4.	Menguraikan tabel
tr_elements = doc.xpath('//tr')
col=[]
i=0
for t in tr_elements[0]:
    i+=1
    name=t.text_content()
    print ('%d:"%s"'%(i,name))
    col.append((name,[]))
```

```python
#5.	Mengubah tipe data rangkaian ke bentuk tuple
for j in range(1,len(tr_elements)):
    T=tr_elements[j] 
    if len(T)!=10:
        break
    i=0

    for t in T.iterchildren():
        data=t.text_content() 
        if i>0:
            try:
                data=int(data)
            except:
                pass
        col[i][1].append(data)
        i+=1
```

```python
#6.	Mengetahui jumlah baris
[len(B) for (title,B) in col]
```

```python
#7.	Mengubah tipe data rangkaian ke dictionary lalu membuat dataframe dari dictionary
Dict={title:column for (title,column) in col}
data=pd.DataFrame(Dict)
```

```python
#12.	Visualisasi scatter plot antara HP dengan Defense

data.plot(x = "HP", y = "Defense", style = "^")
plt.title("Pengaruh HP terhadap Defense")
plt.xlabel("Darah Karakter")
plt.ylabel("Defense Karakter")
plt.show()
```
